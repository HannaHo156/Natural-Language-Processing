## 1. Sentiment Analysis

Use the [newsapi](https://newsapi.org/) to pull the latest news articles for Bitcoin and Ethereum and create a DataFrame of sentiment scores for each coin.

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from path import Path
from dotenv import load_dotenv
from newsapi import NewsApiClient

import nltk as nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

%matplotlib inline
```

```python
# Read your api key environment variable
load_dotenv()
api_key = os.getenv("NEWS_API_KEY")
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch the Bitcoin news articles
btc_news = newsapi.get_everything(
    q="bitcoin",
)
```

```python
# Show the total number of news
btc_news["totalResults"]
```

```python
# Fetch the Ethereum news articles
eth_news = newsapi.get_everything(
    q="ethereum",
)
# Show the total number of news
eth_news["totalResults"]
```

```python
# Create a function of dataframe to store all features of news
def create_df(coinnews):
    articles = []
    for article in coinnews:
        try:
            title = article["title"]
            description = article["description"]
            text = article["content"]
            date = article ["publishedAt"]
            
            articles.append({
                "title":title,
                "description" : description,
                "text" : text,
                "date" : date,
            })
        except AttributeError:
            pass
    return pd.DataFrame(articles)
```

```python
# Create the Bitcoin sentiment scores DataFrame
btc_news_df = create_df(btc_news["articles"])
```

```python
# Create the Ethereum sentiment scores DataFrame
eth_news_df = create_df(eth_news["articles"])
```

```python
# Sentiment calculation based on compound score
def get_sentiment(score):
    result = 0
    if score >= 0.05: #positive
        result = 1
    if score <= 0.05: #negative
        result = -1
    return result
```

```python
# Sentiment scores dictionaries for bitcoin
title_sent1 = {
    "title_compound": [],
    "title_pos": [],
    "title_neu": [],
    "title_neg": [],
    "title_sent": [],
}
text_sent1 = {
    "text_compound": [],
    "text_pos": [],
    "text_neu": [],
    "text_neg": [],
    "text_sent": [],
}
```

```python
# Create funtion to get sentiment for the text and the title
for index, row in btc_news_df.iterrows():
    try:
        # Sentiment scoring with VADER
        title_sentiment1 = analyzer.polarity_scores(row["title"])
        title_sent1["title_compound"].append(title_sentiment1["compound"])
        title_sent1["title_pos"].append(title_sentiment1["pos"])
        title_sent1["title_neu"].append(title_sentiment1["neu"])
        title_sent1["title_neg"].append(title_sentiment1["neg"])
        title_sent1["title_sent"].append(get_sentiment(title_sentiment1["compound"]))

        text_sentiment1 = analyzer.polarity_scores(row["text"])
        text_sent1["text_compound"].append(text_sentiment1["compound"])
        text_sent1["text_pos"].append(text_sentiment1["pos"])
        text_sent1["text_neu"].append(text_sentiment1["neu"])
        text_sent1["text_neg"].append(text_sentiment1["neg"])
        text_sent1["text_sent"].append(get_sentiment(text_sentiment1["compound"]))
    except AttributeError:
        pass
```

```python
# Attaching sentiment columns to the News DataFrame
title_sentiment1 = pd.DataFrame(title_sent1)
text_sentiment1 = pd.DataFrame(text_sent1)
btc_sent_df = btc_news_df.join(title_sentiment1).join(text_sentiment1)
```

```python
# Describe the Bitcoin Sentiment
btc_sent_df.describe()
```

```python
# Sentiment scores dictionaries for Etherum
title_sent2 = {
    "title_compound" : [],
    "title_pos" : [],
    "title_neg" : [],
    "title_neu" : [],
    "title_sent" :[]} #sentiment

text_sent2 = {
    "text_compound" : [],
    "text_pos" : [],
    "text_neg" : [],
    "text_neu" : [],
    "text_sent" :[]} #sentiment
for index, row in eth_news_df.iterrows():
    try:
        # Sentiment scoring with VADER
        title_sentiment2 = analyzer.polarity_scores(row["title"])
        title_sent2["title_compound"].append(title_sentiment2["compound"])
        title_sent2["title_pos"].append(title_sentiment2["pos"])
        title_sent2["title_neu"].append(title_sentiment2["neu"])
        title_sent2["title_neg"].append(title_sentiment2["neg"])
        title_sent2["title_sent"].append(get_sentiment(title_sentiment2["compound"]))

        text_sentiment2 = analyzer.polarity_scores(row["text"])
        text_sent2["text_compound"].append(text_sentiment2["compound"])
        text_sent2["text_pos"].append(text_sentiment2["pos"])
        text_sent2["text_neu"].append(text_sentiment2["neu"])
        text_sent2["text_neg"].append(text_sentiment2["neg"])
        text_sent2["text_sent"].append(get_sentiment(text_sentiment2["compound"]))
    except AttributeError:
        pass

# Attaching sentiment columns to the News DataFrame
title_sentiment2 = pd.DataFrame(title_sent2)
text_sentiment2 = pd.DataFrame(text_sent2)
eth_sent_df = eth_news_df.join(title_sentiment2).join(text_sentiment2)
```

```python
# Describe the Ethereum Sentiment
eth_sent_df.describe()
```

###   Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word.
2. Remove Punctuation.
3. Remove Stopwords.

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re

# Code to download corpora
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Instantiate the lemmatizer
lemmatizer = WordNetLemmatizer()
# Create a list of stopwords
sw = set(stopwords.words('english'))
# Expand the default stopwords list if necessary
regex = re.compile("[^a-zA-Z ]")
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    # Remove the punctuation from text
    re_clean = regex.sub('', text)
    # Create a tokenized list of the words
    words = word_tokenize(re_clean)
    # Lemmatize words into root words
    lem = [lemmatizer.lemmatize(word) for word in words]
    # Convert the words to lowercase
    tokens = [word.lower() for word in lem if word.lower() not in sw]
    # Remove the stop words   
    return tokens
```

```python
# Create a new tokens column for Bitcoin
token = []
btc_token_list = btc_news_df[["title","text"]].copy()
for text in btc_token_list["text"]:
    token_btc = tokenizer(text)
    token.append(token_btc)
btc_token_list["token"] = token
btc_token_list.head(3)
```

```python
# Create a new tokens column for Ethereum
token1 = []
eth_token_list = eth_news_df[["title","text"]].copy()
for text in eth_token_list["text"]:
    token_eth = tokenizer(text)
    token1.append(token_eth)
eth_token_list["token"] = token1
eth_token_list.head(3)
```

### NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
#Create a processfing function for N-Fram
def bigram_counter(corpus):
    big_string = " ".join(corpus)
    processed = tokenizer(big_string)
    bigrams = ngrams(processed, n=2)
    return pd.DataFrame(Counter(bigrams).most_common(10),columns = ["words", "counts"])
```

```python
# Generate the Bitcoin N-grams where N=2
bigram_counter(btc_news_df.text)
```

```python
# Generate the Ethereum N-grams where N=2
bigram_counter(eth_news_df.text)
```

```python
# Function token_count generates the top 10 words for a given coin
def token_count(tokens, N):
    """Returns the top N tokens from the frequency count"""
    return pd.DataFrame(Counter(tokens).most_common(N), columns = ["words", "counts"])
```

```python
# Generate list of all token of Bitcoin
btc_corpus = [item for sublist in token for item in sublist]
# Use token_count to get the top 10 words for Bitcoin
token_count(btc_corpus, 10)
```

```python
# Generate list of all token of Bitcoin
eth_corpus = [item for sublist in token1 for item in sublist]
```

```python
# Use token_count to get the top 10 words for Ethereum
token_count(eth_corpus, 10)
```
